# 6. **Sales Forecasting:**
   - *Problem Statement:* "Accurate sales forecasting is critical for proactive decision-making, resource allocation, and inventory management. The challenge lies in developing models that can predict sales trends with precision, considering various influencing factors such as market dynamics, seasonality, and emerging consumer trends."

```python
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
from math import sqrt



# Create a time series DataFrame
time_series_data = dataset.groupby('SalesDate')['SalesPrice'].sum().reset_index()

# Set 'SalesDate' as the index
time_series_data = time_series_data.set_index('SalesDate')

# Visualize the time series data
plt.figure(figsize=(12, 6))
time_series_data['SalesPrice'].plot(label='Actual Sales')
plt.title('Sales Over Time')
plt.xlabel('SalesDate')
plt.ylabel('Sales Price')
plt.legend()
plt.show()


# Split the data into training and testing sets
train_size = int(len(time_series_data) * 0.8)
train, test = time_series_data.iloc[:train_size], time_series_data.iloc[train_size:]

# Fit SARIMA model
order = (1, 1, 1)  # Order parameters (p, d, q)
seasonal_order = (1, 1, 1, 12)  # Seasonal order parameters (P, D, Q, S)
model = SARIMAX(train, order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
fit_model = model.fit(disp=False)

# Forecast sales
forecast = fit_model.get_forecast(steps=len(test))
forecast_mean = forecast.predicted_mean

# Calculate RMSE (Root Mean Squared Error)
rmse = sqrt(mean_squared_error(test['SalesPrice'], forecast_mean))

# Visualize the forecast
plt.figure(figsize=(12, 6))
time_series_data['SalesPrice'].plot(label='Actual Sales')
forecast_mean.plot(label='Sales Forecast', linestyle='dashed', color='orange')
plt.title(f'Sales Forecasting with SARIMA (RMSE: {rmse:.2f})')
plt.xlabel('SalesDate')
plt.ylabel('Sales Price')
plt.legend()
plt.show()
```
